```python
import pandas as pd
import numpy as np
df = pd.read_csv("crop.csv")
```

```python
crop_dict = df["#1 yield"].value_counts().to_dict()
crop_dict
```

```python
dict={}
for i in crop_dict.keys():
  if(crop_dict[i]>30):
    dict[i] = crop_dict[i]
dict
```

```python
def f(x):
  if x in dict.keys():
    return x
  else:
    return np.nan
df["best_crop"] = df["#1 yield"].apply(f)
df

```

```python
df.drop(columns = ["#1 yield","#2 yield","#3 yield"],inplace=True)
```

```python
df.dropna(inplace=True)
df['Soil_type'] = df["SOIL TYPE"]
df.drop(columns = ["SOIL TYPE"],inplace = True)
```

```python
df["best_crop"].value_counts()
```

```python
len(df)
```

```python
soil_dict = df["Soil_type"].value_counts().to_dict()
soil_dict
soil = {}
for i in soil_dict.keys():
  if(soil_dict[i]>50):
    soil[i] = soil_dict[i]
soil
```

```python
def f(x):
  if x in soil.keys():
    return x
  else:
    return np.nan
df["Soil_type"] = df["Soil_type"].apply(f)
df.dropna(inplace = True)
df["Soil_type"].value_counts()
```

```python
df_final = df.drop(columns = ['Unnamed: 0', 'Year', 'State Name', 'Dist Name', '#1 revenue',
         '#2 revenue', '#3 revenue'])
```

```python
df_final
```

```python
df_final_X = df_final.drop(columns = ["best_crop"])
```

```python
X = pd.get_dummies(df_final_X,drop_first = True).values
y = df_final['best_crop'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3 ,random_state=20)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.linear_model import LogisticRegression
model_params={
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{'n_estimators':range(1,6)}
    },
    'logistic_regression' : {
        'model': LogisticRegression(multi_class='auto'),
        'params': {'penalty':['l1','l2'],
                   'solver':["liblinear",'lbfgs'],
                   'C': [1,5,10]
        }
    },
    'naive_bayes_gaussian': {
        'model': GaussianNB(),
        'params': {}
    },

    "MultinomailNB":{
        'model':MultinomialNB(),
        'params':{}

    },


    "decision_tree":{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion':('gini', 'entropy', 'log_loss')
        }
    }


}
```

```python
earr,earr1,earr2=[],[],[]
dict={
        "model":earr,
        "Params":earr1,
        "Score":earr2
    }
from sklearn.model_selection import RandomizedSearchCV

for model_name,mp in model_params.items():
    clf =RandomizedSearchCV( mp['model'],mp['params'],cv=3,return_train_score=False)
    clf.fit(X_train,y_train)
    earr.append(model_name)
    earr1.append(clf.best_params_)
    earr2.append(clf.best_score_)
```

```python
df=pd.DataFrame(dict,columns=["model","Params","Score"])
```
