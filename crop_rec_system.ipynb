```python
L=list(df_prices.columns)
L_pri=L[3:]
L_rev=["Year","State Name","Dist Name"]
for i in L_pri:
  L_rev.append(i.split("HARVEST PRICE (Rs per Quintal)")[0] + "Revenue")
L_rev
```

```python
df_rev=pd.DataFrame(columns=L_rev)
df_rev
```

```python
dict1=dict(df_yield.iloc[0])
dict2=dict(df_prices.iloc[0])
lol=list(dict1.values())[3:]
lmao=list(dict2.values())[3:]
print(lol)
print(lmao)
```

```python
df_rev2=pd.merge(df_yield,df_prices, how='inner', on=['Year', 'State Name', 'Dist Name'])
df_rev2
```

```python
for i in range(len(df_rev2)):
  dict1=dict(df_rev2.iloc[i])
  L1=list(dict1.values())[3:]
  append_list=[dict1['Year'],dict1['State Name'],dict1['Dist Name']]
  for j in range(13):
    val=L1[j]*L1[j+13]
    if val>0:
      append_list.append(val/100)
    else:
      append_list.append(0)
  df_rev.loc[len(df_rev.index)]=append_list
df_rev
```

```python
rank_1_crop=[]
rank_2_crop=[]
rank_3_crop=[]
for i in range(len(df_rev)):
  my_dict = dict(df_rev.iloc[i])
  values=list(my_dict.values())[3:]
  sort = sorted(values,reverse=True)
  ans1=0
  ans2=0
  ans3=0
  for i in my_dict.keys():
    if(ans1 and ans2 and ans3):
      break
    if(my_dict[i]==sort[0] and ans1==0):
      rank_1_crop.append((i.split(" Revenue")[0]))
      ans1=1
      continue
    if(my_dict[i]==sort[1] and ans2==0):
      rank_2_crop.append(i.split(" Revenue")[0])
      ans2=1
      continue
    if(my_dict[i]==sort[2] and ans3==0):
      rank_3_crop.append(i.split(" Revenue")[0])
      ans3=1
      continue
```

```python
df_rev['Rank 1 crop']=rank_1_crop
df_rev['Rank 2 crop']=rank_2_crop
df_rev['Rank 3 crop']=rank_3_crop
col=list(df_rev.columns)
col=col[3:-3]
df_rev.drop(columns=col,inplace=True)
df_rev
```

```python
df_tmin=pd.read_csv("min temp.csv")
df_tmax=pd.read_csv("max temp.csv")
```

```python
df_tmax.drop(columns=['Dist Code','State Code'], inplace= True)
df_tmin.drop(columns=['Dist Code','State Code'], inplace= True)
```

```python
avg_max=[]
for i in range(len(df_tmax)):
  L=list(dict(df_tmax.iloc[i]).values())
  a=L[3:]
  avg=sum(a)/len(a)
  avg=int(100*avg)/100
  avg_max.append(avg)
df_tmax["Avg Max Temp"]=avg_max
col=list(df_tmax.columns)[3:-1]
df_tmax.drop(columns=col,inplace=True)
df_tmax
```

```python
avg_min=[]
for i in range(len(df_tmin)):
  L=list(dict(df_tmin.iloc[i]).values())
  a=L[3:]
  avg=sum(a)/len(a)
  avg=int(100*avg)/100
  avg_min.append(avg)
df_tmin["Avg Min Temp"]=avg_min
col=list(df_tmin.columns)[3:-1]
df_tmin.drop(columns=col,inplace=True)
df_tmin
```

```python
df_temp=pd.merge(df_tmax,df_tmin,how="inner",on=["State Name","Dist Name","Year"])
```

```python
df_rev_d=pd.merge(df_rev_c,df_temp,how="inner",on=["State Name","Dist Name","Year"])
```
